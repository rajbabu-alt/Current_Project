```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"/kaggle/input/datasets/anmolkumar/analytics-vidhya-loan-prediction/train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.drop(['Loan_ID'], axis=1)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
```

```python
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
df['Property_Area'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df['Married'] = df['Married'].map({'Yes': 1, 'No': 0})
df['Self_Employed'] = df['Self_Employed'].map({'Yes': 1, 'No': 0})
df['Loan_Status'] = df['Loan_Status'].map({'Y':1, 'N':0})
df['Education'] = df['Education'].map({'Graduate': 1, 'Not Graduate': 0})
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['Property_Area'])
```

```python
df = pd.get_dummies(df, columns=['Dependents'])
```

```python
df.head()
```

```python
df['Property_Area_Rural'] = df['Property_Area_Rural'].astype(int)
df['Property_Area_Semiurban'] = df['Property_Area_Semiurban'].astype(int)
df['Property_Area_Urban'] = df['Property_Area_Urban'].astype(int)
```

```python
df['Dependents_0'] = df['Dependents_0'].astype(int)
df['Dependents_1'] = df['Dependents_1'].astype(int)
df['Dependents_2'] = df['Dependents_2'].astype(int)
df['Dependents_3+'] = df['Dependents_3+'].astype(int)
```

```python
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python
df['Income_Loan_Ratio'] = df['TotalIncome'] / df['LoanAmount']
```

```python
df['LoanAmount_log'] = np.log(df['LoanAmount'] + 1)
```

```python
df = df.drop(['ApplicantIncome'], axis=1)
df = df.drop(['CoapplicantIncome'], axis=1)
df = df.drop(['TotalIncome'], axis=1)
df = df.drop(['LoanAmount'], axis=1)
```

```python
df.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
log = LogisticRegression(max_iter=1000, random_state=23)
log_score = cross_val_score(log, X_scaled, y, cv=5, scoring='accuracy')
```

```python
tree = DecisionTreeClassifier(max_depth=5, random_state=23)
tree_score = cross_val_score(tree, X, y, cv=5, scoring='accuracy')
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=23)
forest_score = cross_val_score(forest, X, y, cv=5, scoring='accuracy')
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn_score = cross_val_score(knn, X_scaled, y, cv=5, scoring='accuracy')
```

```python
print("model_comparision")
print(f"LogisticRegression: {log_score.mean():.3%}")
print(f"DecisionTree: {tree_score.mean():.3%}")
print(f"RandomForest: {forest_score.mean():3%}")
print(f"knn : {knn_score.mean():3%}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}
```

```python
grid_search = GridSearchCV(LogisticRegression(max_iter=1000),param_grid, cv=5)
grid_search.fit(X_scaled, y)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
```
